```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
TICKER = 'SPY'
WINDOW = 200

def get_data():
    df = yf.download(TICKER)
    df.columns = df.columns.get_level_values(0)

    # add the moving average column
    df['MA'] = df['Close'].rolling(WINDOW).mean()

    # plot the information
    plt.plot(df['Close'])
    plt.plot(df['MA'])
    plt.title(f'{TICKER} Close Price With {WINDOW} Moving Average')

    return df.dropna()

def main():
    df = get_data()
    return df

main()
```
